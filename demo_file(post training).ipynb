## Check the reconstruction performance

```python
intermediate_layer_model = Model(inputs=model_str.input,
                                 outputs=model_str.output)
########depends on which trials are needed to be tested########################
imageout=np.zeros((len(image_name)*len(ff[image_name[0]][Name]),img_x,img_y,img_z))
laaa=np.zeros((len(image_name)*len(ff[image_name[0]][Name]),img_x,img_y,img_z))


labelout=np.zeros((len(label_name)*len(ff[image_name[0]][Name]),sdim))
lbbb=np.zeros((len(label_name)*len(ff[image_name[0]][Name]),sdim))

Name=list(ff[image_name[0]].keys())[0]

idx={i:[] for i in range (len(label_name)*len(ff[image_name[0]][Name]))}
z=0
trails=list(ff[image_name[0]].keys())[:1]### test only one trial, we have only one trial in the example dataset


for i in range(len(image_name)):

    for j in trails:
        for k in range (len(ff[image_name[0]][Name])):
            idx[z].append(i)
            idx[z].append(j)
            idx[z].append(k)

            z+=1

z=0
ind=idx
for alp in range (len(label_name)*len(ff[image_name[0]][Name])):
    i=ind[alp][0]
    j=ind[alp][1]
    k=ind[alp][2]
    
    im=np.rollaxis(np.expand_dims(np.array(ff[image_name[i]][j][k]),axis=0).T,3,0)
    lab=np.array(ff[label_name[i]][j][k])
    lab=lab.reshape(1,-1)
    intermediate_output,pred_label = intermediate_layer_model.predict([np.rollaxis(im,2,1)/255,lab])#
    imageout[z]=intermediate_output
    labelout[z]=pred_label
    lbbb[z,:]=lab
    laaa[z,:]=np.rollaxis(im,2,1)/255

    z+=1
print('Mse_error for each frame:',np.mean(((imageout-laaa))**2))
from sklearn.metrics import r2_score

for i in range (5):
    r=r2_score(labelout[:,i],lbbb[:,i])
    
    print('r2 for label reconstruction:'+str(i)+'=',r)
```
